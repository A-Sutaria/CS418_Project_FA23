```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# # Time values provided from dataset after cleaning from ZHVF
x_values_str = ["10/31/23", "12/31/23", "9/30/24"]

# Convert the x-values to datetime objects
x_values = [datetime.strptime(date_str, "%m/%d/%y") for date_str in x_values_str]

# Price values provided from dataset after cleaning from ZHVF
y_values = [0.3, 0.7, 2.2]

# Sort the data based on x-values
sorted_indices = sorted(range(len(x_values)), key=lambda k: x_values[k])
x_values = [x_values[i] for i in sorted_indices]
y_values = [y_values[i] for i in sorted_indices]

# Figure size
fig, ax = plt.subplots(figsize=(10, 6))

 

# Plot the graph
ax.plot(x_values, y_values, marker='o', linestyle='-', linewidth=0.5)

# x-axis as dates
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%y'))

# Labels and Title
plt.xlabel('Date')
plt.ylabel('Price Multiplier')
plt.title('Forecast of price in ZHVF')



# Show 
plt.tight_layout()
plt.show()
```
